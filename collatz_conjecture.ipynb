### 3n + 1

Maybe try to find patterns? I'm not a mathematician and I just want to try this out.

- pick a number `N`
- if N is even -> `N = N/2`
- if N is odd -> `N = 3N + 1`

Assumption is that N should grow but it gets smaller until it reaches the `4 -> 2 -> 1` loop

```python
def experiment(n):
    """Actual loop"""
    i = 0 # Number of steps taken
    e = 0 # Number of even numbers passed
    o = 0 # Number of odd numbers passed
    
    prev_flag = n % 2 == 0
    
    while n != 1:
        i += 1
        if n % 2 == 0:
            n = n / 2
            e += 1
        else:
            n = (3 * n) + 1
            o += 1
            
#     print(f"Steps: {i}, Even: {e}, Odd: {o}")
    return i, e, o
```

```python
features = []
for i in range(1, 500):
    steps, even, odd = experiment(i)
    features.append((i, steps, even, odd))
    
```
