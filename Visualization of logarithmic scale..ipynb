```python
from matplotlib import pyplot as plt
from typing import List
```

```python
def generate_numbers(n: int, base: int = 2) -> List[int]:
    """
    Generate N element array of the power of 2.
    """
    
    return [base ** x for x in range(1, n)]
```

```python
def show_plot(arr: list):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 3))
    ax1.plot(arr, color="blue", lw=2)
    ax1.set_yscale("log")
    ax1.title.set_text("Logarithmic")
    ax2.plot(arr, color="blue", lw=2)
    ax2.title.set_text("Non Logarithmic")
    plt.show()
```

```python
bases = [1.01, 1.02, 1.03, 1.04, 1.05, 1.1, 1.2, 1.3, 1.4, 1.5, 2, 2.5, 3, 4, 5, 10]
for b in bases:
    print(f"BASE: {b}")
    arr = generate_numbers(100, b)
    show_plot(arr)
```
